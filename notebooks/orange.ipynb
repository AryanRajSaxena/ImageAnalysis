```python
# Importing Image from PIL package
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
import pandas as pd
```

```python
def load_image(image_path):
    im = Image.open(image_path)
    im =im.resize((256,256))
    crop_box = (100, 100, 150, 150)
    img_cropped = im.crop(crop_box)
    return [img_cropped, im]

def show_image(image):
    img_resized = np.array(image)
    plt.grid()
    plt.imshow(img_resized)
    plt.show()

def color(image):
    redness = []
    for i in range(0,image.size[0]):
        for j in range(0,image.size[1]):
            if image.getpixel((i,j))[0] in range(180,255) and image.getpixel((i,j))[1] in range(100,165) and image.getpixel((i,j))[2] in range(0,50):
                redness.append((image.getpixel((i,j))[0] + image.getpixel((i,j))[1] + image.getpixel((i,j))[2])/3)

    avg_redness = sum(redness) / (len(redness)+1)
    return avg_redness
def files(image_dir,image_files):
    temp = []
    for image_file in image_files:
        # Construct the full image path
        image_path = os.path.join(image_dir, image_file)
        img = load_image(image_path)
        temp.append(color(img[1]))
    return max(temp) #specify operation to be performed on the 3 images color values
    # return (((sum(temp)/len(temp))/255)*100)
```
