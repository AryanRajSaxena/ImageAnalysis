```python
rgb = [255, 165, 0]  # RGB values for orange

# Normalize the RGB values to the [0, 1] range, which is the format matplotlib accepts.
rgb_normalized = [x / 255 for x in rgb]

# Create a figure and a subplot
fig, ax = plt.subplots(1, 1, figsize=(2, 2),
                       subplot_kw=dict(xticks=[],
                                       yticks=[],
                                       frame_on=False))

# Set the color of the subplot to the RGB color
ax.set_facecolor(rgb_normalized)

plt.show()
```
